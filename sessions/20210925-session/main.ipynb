## Nombrar los ejes

```python
dias = np.arange(1,8)
moscas_pancho = np.array([10,8,3,4,11,2,3])
moscas_panfilo = np.array([5,5,6,4,20,10,9])

plt.plot(dias, moscas_pancho, label="Pancho")
plt.plot(dias, moscas_panfilo, label="Panfilo")
plt.xlabel("Días")
plt.ylabel("Moscas")
plt.title("Nutrición del ajolote")
plt.grid(True)
plt.legend()
```
