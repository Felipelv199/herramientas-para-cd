# Fusión de muchos a uno

```python
pd.merge(df1, df2, on="llave")
```

```python
df3 = DataFrame({"llave1": list('bbacaab'), 'valor1': range(7)})
df4 = DataFrame({'llave2': ['a', 'b', 'd'], 'valor2': [10, 11, 12]})
display(df3)
df4
```

```python
pd.merge(df3, df4, left_on="llave1", right_on="llave2")
```

# Tipos de fuciones

* _inner_: Combinanciones que aparecen en ambas tablas _default_
* _left_: Combinaciones con llaves que aparecen en la primer tabla
* _right_: Combinaciones con llaves que aparecen en la segunda tabla
* _outer_: Todas las posibles combinaciones

```python
display(df1)
```

```python

pd.merge(df1, df2, how="inner")
pd.merge(df1, df2, how="left")
pd.merge(df1,df2,how="right")
pd.merge(df1, df2, how="outer")
```

```python
df5 = DataFrame({"llave": list('bbacaab'), 'valor1': range(7)})
df6 = DataFrame({'llave': ['a', 'b', 'd', 'a', 'b'], 'valor2': [10, 11, 12, 13, 14]})
display(df5)
display(df6)
```

```python
display(pd.merge(df5, df6, how="inner"))
display(pd.merge(df5, df6, how="left"))
display(pd.merge(df5, df6, how="right"))
display(pd.merge(df5, df6, how="outer"))
```

```python
df7 = pd.DataFrame({
    'llave1': ['uno', 'dos', 'uno'],
    'llave2': ['perro', 'perro', 'gato'],
    'valor1': [10, 20, 30]
})
df8 = pd.DataFrame({
    'llave1': ['uno', 'uno', 'dos', 'dos'],
    'llave2': ['perro', 'perro', 'perro', 'gato'],
    'valor2': [40, 60, 70, 80]
})
display(df7)
display(df8)
```

```python
pd.merge(df7, df8, on=["llave1"], how="outer")
```
