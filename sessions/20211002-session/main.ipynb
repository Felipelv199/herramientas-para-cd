## Modificar los parámetros para una sola gráfica en específico

```python
x=[1,2,3,4,5]
y=[3,2,1,4,5]
fig = plt.figure() ## Regresa la figura para cambiar sus parámetros
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,y)
```

```python
fig = plt.figure()
(ax1,ax2),(ax3,ax4)=fig.subplots(2,2)
ax1.plot([1,2,3], color="red")
ax2.plot([3,2,1], color="red")
ax3.plot([4,4,4], color="red")
ax4.plot([1, 0, 1], color="red")
fig.tight_layout()
```

```python
rows = 2
cols = 3

fig, axes = plt.subplots(rows, cols)
for i in range(rows):
    for j in range(cols):
        axes[i][j].plot(np.arange(10),"x")

fig.tight_layout()
```

```python
fig = plt.figure(figsize=(10,5))
grid = fig.add_gridspec(3,3)
ax1 = fig.add_subplot(grid[0, :])
ax2 = fig.add_subplot(grid[1:, 0:2])
ax3 = fig.add_subplot(grid[1,2])
ax4 = fig.add_subplot(grid[2, 2])

ax1.set_title("Gráfica 1")
ax1.plot([1,3,2,3])
ax2.bar([1,2,3], [4,2,1])
ax3.scatter([1,2,3], [1,2,3])
ax4.pie([8,7,1])
fig.tight_layout()
```

## Diferentes tipos de gráficas

**Lineal**

```python
## Lineal
x = np.arange(-10, 10, 1)
y=x**2
plt.plot(x,y,"x:b")
```

```python
x = np.arange(-10,10,1)
y=x**2
plt.plot(x,y,color="#FF11FF", linestyle="dashdot", linewidth=4, marker="o", markersize=12, markerfacecolor="red", markeredgecolor="black", markeredgewidth=3)
plt.grid(True)
```

¿Cómo personalizar la cuadrícula?

```python
x = np.arange(-10, 10, 1)
y = x**2
plt.plot(x, y, color="#FF11FF", linestyle="dashdot", linewidth=4, marker="o",
         markersize=12, markerfacecolor="red", markeredgecolor="black", markeredgewidth=3)

# Orientación de los labels
# plt.xticks(x, rotation="vertical")
plt.xticks(x, rotation=45)

plt.yticks(np.arange(0,110,10))
plt.grid(True)
```
