```python
import matplotlib.pyplot as plt

x=[8,4,3,1]
plt.pie(x)
plt.show()
```

```python
labels = ["cerezo", "manzano", "aguacate", "pino"]
plt.pie(x, labels=labels)
plt.show()
```

```python
from random import random


def r_color():
    return (random(), random(), random(), .5)

explode = [0,0,0,.2]
colores = [r_color() for i in x]
plt.pie(x, labels=labels, explode=explode, shadow=True, autopct="%1.1f%%", colors=colores)
plt.show()
```
