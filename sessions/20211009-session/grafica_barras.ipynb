```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,3,4,5,6,7,9])
y = np.array([4,7,2,4,7,8,3])

x2 = np.array([2,4,6,8,10])
y2 = np.array([5,6,2,6,2])

plt.bar(x,y, label="Barras 1")
plt.bar(x2, y2, label="Barras 2")

plt.legend()
plt.show()
```

# ¿Como evitar que una gráfica se sobreponga a la otra?

```python
plt.bar(x, y, label="Barras 1", color=(0,0,1,.5))
plt.bar(x2, y2, label="Barras 2", color=(0,1,0,.5))

plt.xticks(np.arange(1,11))

plt.legend()
plt.show()
```

# Etiquetar las diferentes barras con un nombre en específico

```python
plt.bar(x, y, label="Barras 1", color=(0, 0, 1, .5))
plt.bar(x2, y2, label="Barras 2", color=(0, 1, 0, .5))

xLables = "ABCDEFGHIJ"
plt.xticks(np.arange(1, 11), labels=xLables)

plt.legend()
plt.show()
```

# Utlizacion de una escala de colores para la grafica de barras

```python
from matplotlib import cm

cmap = cm.get_cmap("plasma")

x1 = np.arange(1,20)
y1 = np.arange(1,20)

colores = [cmap(c) for c in y1/max(y1)]

plt.bar(x1, y1, color=colores)
plt.show()
```

# Apilar diferentes gráficas de barras

```python
x = np.arange(1, 20)
y1 = np.arange(1, 20)

y2 = np.arange(19, 0, -1)
y3 = np.random.randint(1,10,19)

plt.bar(x,y1)
plt.bar(x,y2, bottom=y1)
plt.bar(x,y3, bottom=y1+y2)
```

# Para barras horizontales utilizar plt.hbar

```python

plt.barh(x, y1)
plt.barh(x, y2, left=y1)
plt.barh(x, y3, left=y1+y2)
```
