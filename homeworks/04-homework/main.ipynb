```python
import csv

file_name = "../../data/titanic.csv"
file = open(file_name)
data = csv.reader(file)
categories = next(data)
passengers = []

for row in data:
    d = dict()
    for k,v in zip(categories, row):
        d[k]=v
    passengers.append(d)
```

# Gr치fica de barras

```python
import matplotlib.pyplot as plt
import math
import numpy as np

ages_pp = dict()

for passenger in passengers:
    passenger_age = passenger["Age"]
    if passenger_age != '':
        passenger_age_f = float(passenger["Age"])
        passenger_age = math.ceil(passenger_age_f)
        if ages_pp.get(passenger_age) == None:
            ages_pp[passenger_age] = 1
        else:
            ages_pp[passenger_age] += 1

ages_pp_list = [[age, ages_pp[age]] for age in ages_pp]
ages_pp_list.sort(key=lambda x: x[0])

x = np.array([f"{age[0]} years" for age in ages_pp_list])
y = np.array([age[1] for age in ages_pp_list])
# Regresa la figura para cambiar sus par치metros
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(1, 1, 1)
ax1.bar(x, y)
plt.xticks(x, rotation=-65)
ax1.set_xlabel("Ages")
ax1.set_ylabel("Number of Passengers")
```

# Gr치fica Pie

```python
gender_pp = dict()

for passenger in passengers:
    passenger_gender = passenger["Sex"]
    if gender_pp.get(passenger_gender) == None:
        gender_pp[passenger_gender] = 1
    else:
        gender_pp[passenger_gender] += 1

gender_pp_list = [[gender, gender_pp[gender]] for gender in gender_pp]

labels = [pg[0] for pg in gender_pp_list]
x = [pg[1] for pg in gender_pp_list]
explode = [0, .05]

plt.pie(x, labels=labels, explode=explode, shadow=True)
```

# Gr치fica Scatter

```python
x=[]
y=[]

for passenger in passengers:
    age = passenger["Age"]
    fare = passenger["Fare"]
    if age != '' and fare != '':
        p_age_f = float(age)
        p_age = math.floor(p_age_f)
        p_fare = float(fare)
        x.append(p_age)
        y.append(p_fare)
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(x, y)
ax1.set_xlabel("Ages")
ax1.set_ylabel("Fares")
```
