# Filtrado de datos
## Hacer un muestreo ejemplo de los datos (imprime algunos de los datos)

```python
import pandas as pd

file_name = "../../data/building-permits.csv"
building_permits_df = pd.read_csv(file_name)
building_permits_df.head()
```

## Enlistar las columnas que forman el dataset

```python
columns_list = list(building_permits_df.columns)
print(columns_list)
```

## Contar cuántos datos faltantes existen por columna

```python
columns_na_sum = dict(building_permits_df.isna().sum())
print(columns_na_sum)
```

## ¿Qué porcentaje de las filas tienen datos faltantes? (Les faltan algunos datos, les faltan todos los datos)

```python
m, n = building_permits_df.shape

columns_na_per = {column: (columns_na_sum[column]/m)*100 for column in columns_na_sum}
print(columns_na_per)
```

## Revisar las columnas: "Street Number Suffix", "Zipcode"

```python
def print_missing_data_percentage(column_name):
    print(
        f"Porcentaje de datos faltantes: {columns_na_per[column_name]:,.2f}%")

print_missing_data_percentage('Street Number Suffix')
print_missing_data_percentage('Zipcode')
```

## Determinar por qué los datos no están (no existe/ no fue registrado)

```python
column_name = 'Street Number Suffix'
print(f"{column_name} percentage of NaN's: {columns_na_per[column_name]:,.2f}%")
column_name = 'Zipcode'
print(f"{column_name} percentage of NaN's: {columns_na_per[column_name]:,.2f}%")
```

### Street Number Suffix

Debido a que el porcentaje de NaN's en esta columna es casi del 100% podemos concluir que no tiene valores.

### Zipcode

Esta columna presenta un porcentaje bajo de NaN's, por esto concluyo que no se registraron unas celdas.

## Reemplazar los valores NAN con los valores que están directamente después de ellos y reemplazar los restantes con ceros

```python
display(building_permits_df.head())
building_permits_df.fillna(method='ffill', limit=1, inplace=True)
building_permits_df.fillna(0, inplace=True)
display(building_permits_df.head())
```

## Punto Extra

Los datos son de San Francisco. La columna de Zipcode está incompleta. ¿Cómo rellenarían los datos que faltan?

```python
columns = ["Street Name", "Street Number","Zipcode"]
columns_values = building_permits_df[columns].values
streets = dict()

for value in columns_values:
    street = f"{value[0]} {value[1]}"
    if streets.get(street) == None:
        streets[street] = value[2]
    elif streets.get(street) == 0 and value[2] != 0:
        streets[street] = value[2]

df_zipcode = building_permits_df[building_permits_df['Zipcode'] == 0]
m, n = df_zipcode.shape

for i in range(m):
    row = df_zipcode.iloc[i]
    street_name = row["Street Name"]
    street_number = row["Street Number"]
    street = " ".join([street_name, str(street_number)])
    df_zipcode.iloc[i, df_zipcode.columns.get_loc('Zipcode')] = streets[street]

display(building_permits_df)
display(building_permits_df[building_permits_df['Zipcode'] == 0].shape)
building_permits_df[building_permits_df['Zipcode'] == 0] = df_zipcode
display(building_permits_df)
display(building_permits_df[building_permits_df['Zipcode'] == 0].shape)
```
